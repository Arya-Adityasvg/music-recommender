```python
feature_cols =['danceability', 'energy', 'valence', 'tempo', 'acousticness', 'loudness', 'speechiness']
X = df[feature_cols]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.neighbors import NearestNeighbors

model= NearestNeighbors(n_neighbors=11, metric='cosine')
model.fit(X_scaled)
```

```python
def recommend(song_name, n=10):
    matches = df[df['track_name'].str.lower() == song_name.lower()]
    
    if matches.empty:
        return "Song not found. Try another one."
    
    idx = matches.index[0]
    distances, indices = model.kneighbors([X_scaled[idx]], n_neighbors=n+1)
    
    similar_indices = indices[0][1:]
    return df.iloc[similar_indices][['track_name', 'artists']]
```

```python
recommend("Blinding Lights")
```

```python
recommend("Perfect")
```
