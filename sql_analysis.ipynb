```python
import pandas as pd
import sqlite3

df=pd.read_csv('data/clean_tracks.csv')
df.shape
```

```python
conn = sqlite3.connect('data/music.db')
df.to_sql('tracks', conn, if_exists='replace', index=False)
```

```python
query ="SELECT track_name, artists, popularity FROM tracks ORDER BY popularity DESC LIMIT 10"
pd.read_sql(query, conn)
```

```python
query = ""
SELECT track_genre, ROUND(AVG(popularity), AS avg_popularity, COUNT(*) AS num_tracks
FROM tracks
GROUP BY track_genre
ORDER BY avg_popularity DESC
```

```python
query = """
SELECT track_genre, ROUND(AVG(popularity), 2) AS avg_popularity, COUNT(*) AS num_tracks
FROM tracks
GROUP BY track_genre
ORDER BY avg_popularity DESC
LIMIT 15
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT 
  CASE 
    WHEN danceability < 0.33 THEN 'Low danceability'
    WHEN danceability < 0.66 THEN 'Medium danceability'
    ELSE 'High danceability'
  END AS danceability_group,
  ROUND(AVG(popularity), 2) AS avg_popularity,
  COUNT(*) AS num_tracks
FROM tracks
GROUP BY danceability_group
ORDER BY avg_popularity DESC
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT explicit, ROUND(AVG(popularity), 2) AS avg_popularity, COUNT(*) AS num_tracks
FROM tracks
GROUP BY explicit
"""
pd.read_sql(query, conn)
```
